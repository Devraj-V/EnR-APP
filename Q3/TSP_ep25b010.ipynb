```python
import math

def tsp_cost_path(dist):
    n=len(dist)
    x=math.inf

    cost=[[x]*n for _ in range(1<<n)]
    parent=[[-1]*n for _ in range(1<<n)]
    cost[1][0]=0  # start at city 0

    for mask in range(1<<n):   #iterate through all possible subsets of cities
        for u in range(n):
            if not (mask & (1<<u)):
                continue
            for v in range(n):
                if mask & (1<<v):
                    continue
                new_mask=mask|(1<<v)
                new_cost=cost[mask][u]+dist[u][v]
                if new_cost < cost[new_mask][v]:
                    cost[new_mask][v]=new_cost
                    parent[new_mask][v]=u

    final_mask=(1<<n)-1
    ans=x
    last_city=-1

    for i in range(n):     #finding best last city
        total_cost=cost[final_mask][i]+dist[i][0]
        if total_cost<ans:
            ans=total_cost
            last_city=i

    path=[]     #path reconstruction
    mask=final_mask
    curr=last_city

    while curr!=-1:
        path.append(curr)
        prev=parent[mask][curr]
        mask=mask^(1<<curr)
        curr=prev

    path.reverse()
    path=[0]+path[1:]+[0]
    return ans,path


# Example input
dist = [
    [0,10,15,20],
    [10,0,35,25],
    [15,35,0,30],
    [20,25,30,0]
]

cost,path=tsp_cost_path(dist)

print("Minimum cost:",cost)
print("Path:",path)
```
